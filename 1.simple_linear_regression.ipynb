```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Print the shape of the data and first row
print(X.shape)
print(X[0])
```

### Create BMI ndarray 

This bellow block of code will create a 2D array that give as the data we want to train it. 

```python
# Extract the column at index 2
X = X[:, 2]
print(X.shape)

# Reshape to a 2D array
X = X.reshape(-1, 1)
print(X.shape)
```

### 2. Split Dataset into Train and Test data

```python
X_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

### 3. Create the Linear Regreation Model

```python
model = linear_model.LinearRegression()
```

### 4. Train the data

```python
model.fit(X_train, y_train)
```

### 5. Predict Using test data

```python
y_pred = model.predict(x_test)
```

6. Evaluate the model

```python
# Create a Scatter plot

plt.scatter(x_test, y_test, color="black")

# PLot the prediction
plt.plot(x_test, y_pred, color="blue", linewidth=3)

# Add labels and a title
plt.xlabel ('Scaled BMIs')
plt.ylabel('Disease progression')
plt.title("A Graph plot showign Diabets Progression Against BMI")

# Draw the plot
plt.show()
```
